TASK 1 - TOP CUISINES

##Determine the top three most common cuisines in the dataset.

```python
#First i will split the cuisines column into lists and explode into separate rows
df['Cuisines'] = df['Cuisines'].str.split(', ')
df_exploded = df.explode('Cuisines')
```

```python
#Count occurrences of each cuisine
cuisine_counts = df_exploded['Cuisines'].value_counts()
cuisine_counts
```

```python
#Top three most common cuisines
top_3_cuisines = cuisine_counts.head(3)
print(f"Top 3 Most Common Cuisines :\n{top_3_cuisines}")
```

##Calculate the percentage of restaurants that serve each of the top cuisines.

```python
#Total number of restaurants
total_restaurants = len(df)
total_restaurants 
```

```python
#Percentage of restaurants that serve each of the top cuisines
top_3_cuisine_percentages = round((top_3_cuisines / total_restaurants) * 100, 2)
print("Percentage of Restaurants Serving Top 3 Cuisines :")
for cuisine, percentage in top_3_cuisine_percentages.items():
    print(f"{cuisine}: {percentage:.2f}%")
```

SUMMARY

The top three most common cuisines in the dataset are as follows:
1. North Indian: 41.56% of restaurants serve this cuisine.
2. Chinese: 28.64% of restaurants serve this cuisine.
3. Fast Food: 20.79% of restaurants serve this cuisine.

INSIGHT - North Indian cuisine dominates the restaurant landscape, with nearly 42% of establishments offering it, showcasing its widespread popularity. Chinese cuisine follows as a strong second, available in about 29% of restaurants, reflecting its appeal across diverse diners. Fast Food, offered by approximately 21% of restaurants, secures its place as a convenient and favored option.

TASK 2 - CITY ANALYSIS

##Identify the city with the highest number of restaurants in the dataset.

```python
#Count restaurants by city
city_counts = df['City'].value_counts()
city_counts
```

```python
#City with the highest number of restaurants
top_city = city_counts.idxmax()
#top_city
top_city_count = city_counts.max()
#top_city_count

print(f"The city with the highest number of restaurants is {top_city}, with {top_city_count} restaurants.")
```

##Calculate the average rating for restaurants in each city.

```python
#Average rating by city
city_avg_ratings = df.groupby('City')['Aggregate_rating'].mean().round(2).sort_values(ascending = False)

print(f"Average ratings for each city:\n {city_avg_ratings}")
```

##Determine the city with the highest average rating.

```python
#City with the highest avg rating
highest_rated_city = city_avg_ratings.idxmax()
highest_rating = city_avg_ratings.max()

print(f"The city with the highest average rating is {highest_rated_city}, with an average rating of {highest_rating:.2f}.")
```

SUMMARY

City with the Most Restaurants --> New Delhi, hosting 5,473 restaurants, dominates the dataset in terms of restaurant count.

City with the Highest Average Rating --> Inner City, with an impressive average rating of 4.90, stands out as the city with the best-rated restaurants.

INSIGHT - New Delhi emerges as the central hub for dining with the highest number of restaurants, indicating its culinary diversity and popularity. However, Inner City takes the lead in quality, reflecting its high standard of dining experiences. The contrast between quantity (New Delhi) and quality (Inner City) suggests varying restaurant density and customer satisfaction trends across cities.

TASK 3 - Price Range Distribution

##Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.

```python
#count the occurrences of each price range
price_counts = df['Price_range'].value_counts().sort_index()

#bar chart to visualize price range distribution
plt.figure(figsize = (8, 6))
sns.barplot(x = price_counts.index, y = price_counts.values, color = 'skyblue')

#Add data labels
for i, count in enumerate(price_counts.values):
    plt.text(i, count, str(count), ha = 'center', va = 'bottom')

#Customizs the chart
plt.title('Price Range Distribution')
plt.xlabel('Price Range')
plt.ylabel('No. of Restaurants')
plt.xticks(ticks = [0, 1, 2, 3], labels = ['Low', 'Medium', 'High', 'Luxury'])
plt.show()
```

##Calculate the percentage of restaurants in each price range category.

```python
#Calculate percentages
price_range_percentages = round(df['Price_range'].value_counts(normalize = True) * 100, 2)

#Display the percentages
print("Percentage of restaurants in each Price Range:")
for price_range, percentage in price_range_percentages.items():
    print(f"{price_range}: {percentage:.2f}%")
```

SUMMARY

The distribution of restaurants by price range is as follows:                                                                                           
Low (Price Range 1): 46.53% of restaurants.                                                                                                             
Medium (Price Range 2): 32.59% of restaurants.                                                                                                          
High (Price Range 3): 14.74% of restaurants.                                                                                                            
Luxury (Price Range 4): 6.14% of restaurants.

INSIGHT - The data shows that nearly half of the restaurants fall under the low-price range, making it the most accessible and preferred choice for customers. The medium range follows with a significant proportion, indicating a balance between affordability and quality. Restaurants in the high and luxury price ranges account for just over 20% of the dataset, highlighting their role as niche offerings catering to premium customers seeking upscale dining experiences. 

TASK 4 - Online Delivery

##Determine the percentage of restaurants that offer online delivery.

```python
#Calculate the percentage of restaurants with online delivery
online_delivery_percentage = round(df['Has_Online_delivery'].value_counts(normalize = True) * 100, 2)
online_delivery_percentage = online_delivery_percentage.astype(str) + '%'

print(f"Percentage of Restaurants Offering Online Delivery : {online_delivery_percentage}")
```

##Compare the average ratings of restaurants with and without online delivery.

```python
#Compare avg ratings of restaurants with and without online delivery
avg_rating_online = df[df['Has_Online_delivery'] == 'Yes']['Aggregate_rating'].mean()
avg_rating_offline = df[df['Has_Online_delivery'] == 'No'] ['Aggregate_rating'].mean()

print(f"Average rating of restaurants with online delivery: {avg_rating_online:.2f}")
print(f"Average rating of restaurants without online delivery: {avg_rating_offline:.2f}")
```

```python
#Creating visual to compare avg ratings
ratings = [avg_rating_online, avg_rating_offline]
categories = ['With Online Delivery', 'Without Online Delivery']

# Creating the bar chart
plt.bar(categories, ratings, color=['green', 'orange'])
plt.title('Average Ratings of Restaurants by Online Delivery Availability')  
plt.xlabel('Online Delivery Availability') 
plt.ylabel('Average Rating')  
plt.show()
```

SUMMARY 

Online Delivery Availability: 25.66% of restaurants offer online delivery, while the remaining 74.34% do not.                                                                  
Average Ratings Comparison: Restaurants with online delivery have an average rating of 3.25, while restaurants without online delivery have a lower average rating of 2.47.

INSIGHT - Restaurants offering online delivery receive higher average ratings (3.25) compared to those without (2.47), indicating a potential correlation between online delivery availability and customer satisfaction. This could be attributed to the convenience and accessibility that online delivery provides, which likely enhances the overall customer experience.
