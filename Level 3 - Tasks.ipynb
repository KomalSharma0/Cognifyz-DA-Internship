TASK 1 - RESTAURANT REVIEWS

##Analyze the text reviews to identify the most common positive and negative keywords.

```python
from collections import Counter
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string 
import re

#Download nltk stopwords
#nltk.download('stopwords')
#nltk.download('punkt')
```

```python
# Preprocess reviews
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    phrases = ['very good', 'not rated']  # Add multi-word phrases to preserve
    for phrase in phrases:
        text = text.replace(phrase, phrase.replace(' ', '_'))  # Replace spaces in phrases
    tokens = word_tokenize(text)  # Tokenize text
    tokens = [word for word in tokens if word not in stopwords.words('english')]  # Remove stopwords
    return tokens

# Apply preprocessing
df['tokens'] = df['Rating_text'].apply(preprocess_text)

# Define categories
positive_reviews = []
negative_reviews = []
unrated_reviews = []

# Separate tokens based on categories
for _, row in df.iterrows():
    tokens = row['tokens']
    rating = row['Aggregate_rating']
    if isinstance(tokens, list):  # Ensure tokens is a list
        if rating >= 3.5:
            positive_reviews.extend(tokens)
        elif 0 < rating < 3.5:
            negative_reviews.extend(tokens)
        elif row['Rating_text'].lower() == 'not rated':
            unrated_reviews.extend(tokens)

# Restore original multi-word phrases
positive_reviews = [word.replace('_', ' ') for word in positive_reviews]
negative_reviews = [word.replace('_', ' ') for word in negative_reviews]
unrated_reviews = [word.replace('_', ' ') for word in unrated_reviews]

# Find most common keywords
positive_keywords = Counter(positive_reviews).most_common(10)
negative_keywords = Counter(negative_reviews).most_common(10)
unrated_keywords = Counter(unrated_reviews).most_common(10)

# Print results
print("Most Common Positive Keywords:", positive_keywords)
print("Most Common Negative Keywords:", negative_keywords)
print("Most Common Unrated Keywords:", unrated_keywords)
```

##Calculate the average length of reviews and explore if there is a relationship between review length and rating.

```python
# Calculate the length of each review in terms of words
df['review_length'] = df['Rating_text'].apply(lambda x: len(word_tokenize(x)))

# Calculate the average length of reviews for the dataset
average_review_length = df['review_length'].mean()
print(f"Average Review Length: {average_review_length:.2f} words")

# Analyze the relationship between review length and aggregate rating
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Aggregate_rating', y='review_length', alpha=0.7, color='blue')
plt.title('Relationship Between Review Length and Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Review Length (Number of Words)')
plt.grid(True)
plt.show()

# Optional: Calculate average review length by rating category
average_length_by_rating = df.groupby('Aggregate_rating')['review_length'].mean().reset_index()
print("Average Review Length by Rating Category:")
print(average_length_by_rating)
```

TASK 2 - VOTES ANALYSIS

##Identify the restaurants with the highest and lowest number of votes.

```python
#Filter out unrated restaurants
rated_restaurants = df[df['Aggregate_rating'] > 0]

#Identify restaurants with the highest and lowest votes
highest_votes = rated_restaurants[rated_restaurants['Votes'] == rated_restaurants['Votes'].max()]
lowest_votes = rated_restaurants[rated_restaurants['Votes'] == rated_restaurants['Votes'].min()]

print("Restaurant with the Highest Number of Votes:")
print(highest_votes[['Restaurant_ID', 'Votes', 'Aggregate_rating']])

print("Restaurant with the Lowest Number of Votes:")
print(lowest_votes[['Restaurant_ID', 'Votes', 'Aggregate_rating']])
```

##Analyze if there is a correlation between the number of votes and the rating of a restaurant.

```python
#Analyze the correlation between Votes and Aggregate Rating
correlation = rated_restaurants[['Votes', 'Aggregate_rating']].corr().iloc[0, 1]

print(f"\nCorrelation Between Votes and Rating: {correlation:.2f}")
```

```python
#Visualize the relationship between Votes and Ratings
plt.figure(figsize = (10, 6))
sns.scatterplot(data = rated_restaurants, x = 'Votes', y = 'Aggregate_rating', alpha = 0.7, color = 'green')
plt.title('Correlation Between Votes and Rating')
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate Rating')
plt.grid(True)
plt.show()
```

TASK 3 - PRICE RANGE vs. ONLINE DELIVERY and TABLE BOOKING

##Analyze if there is a relationship between the price range and the availability of online delivery and table booking.

```python
#Analyze the relationship between Price Range and Online Delivery
online_delivery_by_price = df.groupby('Price_range')['Has_Online_delivery'].value_counts(normalize = True).unstack() * 100

print("Percentage of Online Delivery Availability by Price Range:")
print(online_delivery_by_price)

#Analyze the relationship between Price Range and Table Booking
table_booking_by_price = df.groupby('Price_range')['Has_Table_booking'].value_counts(normalize = True).unstack() * 100
print("Percentage of Table Booking Availability by Price Range:")
print(table_booking_by_price)

#Visualize Price Range vs Online Delivery
plt.figure(figsize = (10, 6))
online_delivery_by_price.plot(kind = 'bar', stacked = True, color = ['lightgreen', 'orange'], alpha = 0.8)
plt.title('Online Delivery Availability by Price Range')
plt.ylabel('Percentage')
plt.xlabel('Price Range')
plt.xticks(rotation = 0)
plt.legend(title = 'Online Delivery', labels = ['No', 'Yes'])
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)
plt.show()

#Visualize Price Range vs Table Booking
plt.figure(figsize = (10, 6))
table_booking_by_price.plot(kind = 'bar', stacked = True, color = ['skyblue', 'purple'], alpha = 0.8)
plt.title('Table Booking availability by Price Range')
plt.ylabel('Percentage')
plt.xlabel('Price Range')
plt.xticks(rotation = 0)
plt.legend(title='Table Booking', labels=['No', 'Yes'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

##Determine if higher-priced restaurants are more likely to offer these services.

```python
# Visualization - Price Range vs. Online Delivery
plt.figure(figsize=(8, 6))
online_delivery_by_price.plot(kind='bar', stacked=True, color=['lightgreen', 'orange'], alpha=0.8)
plt.title('Online Delivery Availability by Price Range')
plt.ylabel('Percentage')
plt.xlabel('Price Range')
plt.xticks(rotation=0)
plt.legend(title='Online Delivery', labels=['No', 'Yes'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Visualization - Price Range vs. Table Booking
plt.figure(figsize=(8, 6))
table_booking_by_price.plot(kind='bar', stacked=True, color=['skyblue', 'purple'], alpha=0.8)
plt.title('Table Booking Availability by Price Range')
plt.ylabel('Percentage')
plt.xlabel('Price Range')
plt.xticks(rotation=0)
plt.legend(title='Table Booking', labels=['No', 'Yes'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
